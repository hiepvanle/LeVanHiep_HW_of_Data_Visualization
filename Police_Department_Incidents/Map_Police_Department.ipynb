```python
import folium
import pandas as pd
```

```python
try:
  from google.colab import files
  uploaded = files.upload()
  file_name = list(uploaded.keys())[0]
except:
  file_name = 'Police_Department_Incidents_-_Previous_Year__2016_.csv'
```

```python
data = pd.read_csv(file_name)
data
```

```python
# Take a sample of 500 records
data_sample = data.sample(n=500, random_state=42)

# Initialize the map centered around San Francisco
smap = folium.Map(location=[37.7749, -122.4194], zoom_start=12)

# Function to add a marker
def add_marker(row):
    folium.Marker(
        location=[row['Y'], row['X']],
        popup=f"{row['Category']}: {row['Descript']}\nLocation: {row['Address']}",
        tooltip=row['Category']
    ).add_to(smap)

# Use apply to add markers
data_sample.apply(add_marker, axis=1)

# Save the map to an HTML file
smap.save("map.html")
smap
```
